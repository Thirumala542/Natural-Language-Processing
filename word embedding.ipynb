### Word Embedding Techniques using Embedding Layer in Keras

```python
#pip install tensorflow
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=["I am a good boy",
      "You are a good girl",
      "A for Apple",
      "I am a Data Scientist",
      "This NLP blog is so easy to understand",
      "Refer my other blogs too",
      "I had written blog on Machine Learning",
      "I had written blog on Data Preprocessing",
      "I had written blog on Statistics",
      "I had written blog on Deep Learning"]
```

```python
sent
```

```python
### Vocabulary size
voc_size=10000
```

#### One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
